```python
with open('input7.txt') as f:
    text = f.readlines()
    text = [line.strip() for line in text]

print(text)
```

```python
line0 = text[0]
print(line0)
```

**Organize the bags in the rules as a dictionary as: `{parent_bag: [child_bag, child_bag]}`**

```python
def extract_bags(line_list):
    bag_names = []
    for i, word in enumerate(line_list):
        if word.isnumeric():
            bag_names.append(line_list[i+1] + " " + line_list[i+2])
    return bag_names
```

```python
extract_bags(line0.split())
```

```python
bags = {}
for line in text:
    line = line.split()
    parent_bag = ' '.join(line[:2])
    bags[parent_bag] = extract_bags(line)
bags
```

**Recursively search the dictionary looking for gold bags**

```python
target = 'shiny gold'
bags_containing_gold = []

def find_gold(parent_bag):
    if len(bags[parent_bag]) > 0:
        # Look in a bag (key) for shiny gold (in value list)
        # If the bag has shiny gold, return True
        if target in bags[parent_bag]:
            return True
        # If not, find_gold() for each bag in the bag
        else:
            for child_bag in bags[parent_bag]:
                return find_gold(child_bag)
```

```python
for bag in bags:
    if find_gold(bag):
        bags_containing_gold.append(bag)
len(set(bags_containing_gold))
```

**Since that isn't working, we'll try it in a super inelegant way that involves running a `for` loop over and over until you get the same number multiple times.**

```python
target = 'shiny gold'
has_gold = []
for bag in bags:
    if target in bags[bag]:
        has_gold.append(bag)
has_gold
```

```python
# There's definitely a better way, but it works!
bag_count = []

for i in range(2):
    for bag in bags:
        for gold_holder in has_gold:
            if gold_holder in bags[bag]:
                has_gold.append(bag)
    bag_count.append(len(set(has_gold)))
while bag_count[-2] < bag_count[-1]:
    for bag in bags:
        for gold_holder in has_gold:
            if gold_holder in bags[bag]:
                has_gold.append(bag)

    bag_count.append(len(set(has_gold)))

bag_count[-1]
```
