### How many individual bags are required inside your single shiny gold bag?

````python
with open('input7.txt') as f:
    text = f.readlines()
    text = [line.strip() for line in text]

print(text)
````

````python
line0 = text[0]
print(line0)
````

**Organize the bags in the rules as nested dictionaries: 
`{parent_bag: {child_bag: amount(int), 
              child_bag, amount(int),
              ...}}```**

````python
def extract_bags(line_list):
    bag_names = {}
    for i, word in enumerate(line_list):
        if word.isnumeric():
            bag_names[(line_list[i+1] + " " + line_list[i+2])] = int(word)
    return bag_names
````

````python
extract_bags(line0.split())
````

````python
bags = {}
for line in text:
    line = line.split()
    parent_bag = ' '.join(line[:2])
    bags[parent_bag] = extract_bags(line)
bags
````

````python
total = []

def count_bags(parent_color):
    parent_dict = bags[parent_color]
    for child_color in parent_dict:
        total.append(parent_dict[child_color])
        for i in range(parent_dict[child_color]):
            count_bags(child_color)
````

````python
count_bags('shiny gold')
sum(total)
````
